```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv(r"C:\Users\sibur\OneDrive\Documents\linear_regression_from_scratch\Salary_Data.csv")
X = ds["YearsExperience"].values
Y = ds["Salary"].values
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
def mean(X):
    return np.sum(X)/len(X)
```

```python
def variance(X):
    mean_value = mean(X)
    return np.sum((X-mean_value)**2)/len(X)
```

```python
def norm(X):
    mean_value = mean(X)
    variance_value = variance(X)
    return (X - mean_value)/np.sqrt(variance_value)
```

```python
X_norm = norm(X)
```

```python
X_norm
```

```python
plt.plot(X_norm,Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_norm = X_norm.reshape(-1,1)
```

```python
reg = LinearRegression().fit(X_norm, Y)
```

```python
reg.score(X_norm, Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(X_norm)
```

```python
y_pred
```

```python
plt.plot(X_norm, Y)
plt.plot(X_norm, y_pred)
```
